```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf

def analyze_and_plot_signal(df, col_name, period=4):
    """
    Performs full time-series diagnosis on a single behavioral column.
    
    Args:
        df: Master DataFrame (must have timestamp index or column)
        col_name: The column to analyze (e.g., 'audio_rms', 'gaze_y')
        period: The 'seasonality' cycle. 
                For 2 FPS (0.5s), period=10 implies a 5-second cycle check.
    """
    print(f"\nüî¨ ANALYZING: {col_name}")
    
    # 1. Preprocessing: Handle missing values (NaNs break statsmodels)
    signal = df[col_name].fillna(method='ffill').fillna(0)
    
    # --- A. DECOMPOSITION (Trend vs Habits) ---
    # Model='additive' because behavioral signals usually stack (Baseline + Action)
    decomposition = seasonal_decompose(signal, model='additive', period=period, extrapolate_trend='freq')
    
    trend = decomposition.trend
    seasonal = decomposition.seasonal
    residual = decomposition.resid

    # --- B. PLOTTING DECOMPOSITION ---
    fig, axes = plt.subplots(4, 1, sharex=True, figsize=(10, 8))
    
    # Original Signal
    axes[0].plot(signal, label='Original', color='black', alpha=0.6)
    axes[0].set_ylabel('Raw Data')
    axes[0].set_title(f'Decomposition of {col_name}')
    axes[0].legend(loc='upper left')

    # Trend (The Arc)
    axes[1].plot(trend, label='Trend', color='blue')
    axes[1].set_ylabel('Trend')
    axes[1].legend(loc='upper left')

    # Seasonality (The Loop)
    axes[2].plot(seasonal, label='Seasonality', color='green')
    axes[2].set_ylabel('Seasonality')
    axes[2].legend(loc='upper left')

    # Residuals (The Spikes/Nerves)
    axes[3].scatter(range(len(residual)), residual, label='Residuals', color='red', s=5, alpha=0.6)
    axes[3].axhline(0, color='black', linestyle='--', linewidth=1)
    axes[3].set_ylabel('Residuals')
    axes[3].legend(loc='upper left')
    
    plt.xlabel('Frame Index')
    plt.tight_layout()
    plt.show()

    # --- C. ACF & PACF PLOTS (The Fidget Detectors) ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # ACF: "If I moved 1 second ago, am I moving now?"
    # High ACF = "Sticky" behavior (Fidgeting loops)
    plot_acf(signal, lags=30, ax=ax1, title=f'Autocorrelation (ACF) - {col_name}')
    
    # PACF: Direct correlation removing intermediate steps
    plot_pacf(signal, lags=30, ax=ax2, title=f'Partial Autocorrelation (PACF) - {col_name}')
    
    plt.show()

    # --- D. AGENT INTERPRETATION GENERATOR ---
    # This converts the plots above into data the LLM can read.
    
    # 1. Trend Direction
    delta = trend.iloc[-1] - trend.iloc[0]
    if delta > 0.05: trend_str = "Rising (Increasing Intensity)"
    elif delta < -0.05: trend_str = "Falling (Decreasing Intensity)"
    else: trend_str = "Stable (Consistent)"
    
    # 2. Nervousness Score (Residual Volatility)
    # High Std Dev in residuals = Erratic, unpredicted spikes
    nervousness_score = round(residual.std(), 4)
    
    # 3. Habit Strength (ACF Lag-1)
    # If Lag-1 is > 0.8, the behavior is very slow/sticky (Postural).
    # If Lag-1 is < 0.3, the behavior is twitchy/random.
    acf_val = acf(signal, nlags=1)[1]
    
    report = {
        "metric": col_name,
        "trend": trend_str,
        "stability_score": nervousness_score, # Lower is better
        "predictability": round(acf_val, 2)
    }
    
    return report

# --- HOW TO RUN IT ---
# Assume 'master_df' is your merged dataframe
analyze_and_plot_signal(av_merged, 'h_ratio')  # Check Volume
# analyze_and_plot_signal(master_df, 'gaze_y')     # Check Head Tilt
```
