## Trying to define statistically robust thresholds

Calculating something called MAD (Median Absolute Distance) used to calculate spread of the data but uses
median instead of mean, will be handy for threshold calculation as it is less affected by the outliers

```python
def calculate_mad(values: np.array) -> float:
    median = np.median(values)
    deviations = np.abs(values - median)
    mad = np.median(deviations)
    return mad.item()
```

```python
type(fa_data['h_ratio'].dropna().values)
```

```python
# testing it
mad = calculate_mad(values=fa_data['h_ratio'].dropna().values)
mad
```

Like we used to calculate z score which was (value - mean)/std now we will calculate something called 
modified z score which is calculated as follows : 0.6745 * (value - median)/mad

```python
CONS = 0.6745
```

```python
# calculating modified z score
median = np.median(fa_data['h_ratio'].dropna().values)
mod_z_score = (CONS * (fa_data['h_ratio'].iloc[0] - median)/mad).item()
mod_z_score
```

Now I will try to create a function that passes a window of data, So when are trying to do is not calculate the mod_z_score for the whole timeline, we will calculate this mod_z_score for this sliding window timelines.

```python
def get_sliding_window(values: np.array, current_index: int, window_len: int) -> np.array:
    start = max(0, current_index - window_len)
    window = values[start : current_index + 1]
    
    return window
```

```python
# testing the function
get_sliding_window(values=fa_data['h_ratio'].dropna().values, current_index=2, window_len=10), len(get_sliding_window(values=fa_data['h_ratio'].dropna().values, current_index=2, window_len=10))
```

So, Now if we calculate mod_z_score it will tell us: How unusual is this value compared to recent behavior?

```python
# let's create a function that calculates these mod_z_scores
def calculate_modified_zscore(value: float, reference_values: np.array) -> float:
    CONS = 0.6745
    median = np.median(reference_values).item()
    mad = calculate_mad(values=reference_values)
    if mad == 0:
        mod_z_score = 0.0
    else:
        mod_z_score = CONS * (value - median)/mad
    return mod_z_score
```

```python
# testing the function
reference = [0.1, 0.12, 0.11, 0.13, 0.10, 0.11]
test_value = 0.25
z = calculate_modified_zscore(test_value, reference)
z
```

Now, We will calculate the EMA (Exponential Moving Average)

```python
def update_ema(currect_value: float, previous_ema: float, alpha: float = 0.3) -> float:
    updated_value = alpha * currect_value + (1-alpha) * previous_ema
    return updated_value
```

```python
# Test:
ema = 0.10
values = [0.12, 0.15, 0.13, 0.14, 0.50]

for v in values:
    ema = update_ema(v, ema, alpha=0.3)
    print(f"Value: {v:.2f}, EMA: {ema:.3f}")
```

Now, There are some critical choices we have to make what window size we should, we can choose adaptive window size which will be inversely proportional to variance of the feature

```python
def calculate_adaptive_window(values: np.array, min_window_size: int = 10, max_window_size: int = 60) -> int:
    
    var = np.var(values)
    nor_var = var / (var + 1)
    
    window_size = max_window_size - nor_var * (max_window_size - min_window_size)
    return int(window_size)
```

```python
# Test:
stable_data = [0.1, 0.11, 0.12, 0.11, 0.10, 0.11]  # Low variance
variable_data = [0.1, 0.5, 0.2, 0.8, 0.15, 70.5]    # High variance

print(f"Stable window: {calculate_adaptive_window(stable_data)}") 
print(f"Variable window: {calculate_adaptive_window(variable_data)}")  
```

#### Putting it all together

```python
def detect_anomaly_at_timestamp(
    feature: np.array,
    current_index: int,
    sensitivity: float = 3.5,
):
    # get the adaptive window size
    win_size = calculate_adaptive_window(values=feature)
    
    # get the values in the window for current timestamp
    window = get_sliding_window(values=feature, current_index=current_index, window_len=win_size)
    
    # get the current value
    current_value = feature[current_index]
    
    # calculate mod z_score
    mod_z_score = calculate_modified_zscore(value=current_value, reference_values=window)
    
    # check anomaly
    is_anomaly = abs(mod_z_score) > sensitivity
    
    return is_anomaly, mod_z_score
```

```python
def merge_windowed(words_df, visual_df, audio_df, window_size=0.5):
    """
    Create fixed time windows and aggregate words
    """
    
    # Get time range
    max_time = max(
        visual_df['Time'].max(),
        audio_df['Time'].max(),
        words_df['end'].max()
    )
    
    # Create time windows
    time_windows = np.arange(0, max_time + window_size, window_size)
    
    result = []
    
    for t in time_windows:
        window_start = t
        window_end = t + window_size
        
        # Find words in this window
        words_in_window = words_df[
            (words_df['start'] >= window_start) & 
            (words_df['start'] < window_end)
        ]
        
        # Combine
        result.append({
            'Time': t,
            'words': words_in_window['text'].tolist(),
            'text_concat': ' '.join(words_in_window['text'].tolist()),
        })
        
    return pd.DataFrame(result)
```
